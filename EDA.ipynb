Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Import and explore dataset

```python
df = pd.read_csv("Insurance claims data.csv")
```

```python
df.head(10)
```

```python
# policy_id                         : Unique identifier for each insurance policy
# subscription_length               : Duration of the insurance subscription period
# vehicle_age                       : Age of the insured vehicle in years
# customer_age                      : Age of the customer
# region_code                       : Encoded identifier of the customer’s region
# region_density                    : Population density of the customer’s region
# segment                           : Vehicle market segment (e.g., economy, mid-range, luxury)
# model                             : Vehicle model name
# fuel_type                         : Type of fuel used by the vehicle
# max_torque                        : Maximum torque produced by the vehicle engine
# max_power                         : Maximum power output of the vehicle engine
# engine_type                       : Type/configuration of the vehicle engine
# airbags                           : Number of airbags installed in the vehicle
# is_esc                            : Indicates whether the vehicle has Electronic Stability Control
# is_adjustable_steering            : Indicates whether the steering wheel is adjustable
# is_tpms                           : Indicates whether the vehicle has a Tire Pressure Monitoring System
# is_parking_sensors                : Indicates whether the vehicle is equipped with parking sensors
# is_parking_camera                 : Indicates whether the vehicle has a parking camera
# rear_brakes_type                  : Type of rear brake system used in the vehicle
# displacement                      : Engine displacement capacity
# cylinder                          : Number of engine cylinders
# transmission_type                 : Type of vehicle transmission (e.g., Manual, Automatic)
# steering_type                     : Type of steering mechanism used
# turning_radius                    : Minimum turning radius of the vehicle
# length                            : Length of the vehicle
# width                             : Width of the vehicle
# gross_weight                      : Gross weight of the vehicle
# is_front_fog_lights               : Indicates whether the vehicle has front fog lights
# is_rear_window_wiper              : Indicates whether the vehicle has a rear window wiper
# is_rear_window_washer             : Indicates whether the vehicle has a rear window washer
# is_rear_window_defogger           : Indicates whether the vehicle has a rear window defogger
# is_brake_assist                   : Indicates whether the vehicle has a brake assist system
# is_power_door_locks               : Indicates whether the vehicle has power door locks
# is_central_locking                : Indicates whether the vehicle has central locking
# is_power_steering                 : Indicates whether the vehicle has power steering
# is_driver_seat_height_adjustable  : Indicates whether the driver seat height is adjustable
# is_day_night_rear_view_mirror     : Indicates whether the rear view mirror supports day/night mode
# is_ecw                            : Indicates whether the vehicle has electronic crash warning
# is_speed_alert                    : Indicates whether the vehicle has a speed alert system
# ncap_rating                       : Safety rating of the vehicle based on NCAP standards
# claim_status                      : Target variable indicating whether an insurance claim was filed
```

```python
features = [
    'is_power_door_locks', 'is_central_locking', 'is_power_steering', 
    'is_driver_seat_height_adjustable', 'is_day_night_rear_view_mirror', 
    'is_ecw', 'is_speed_alert','is_brake_assist'
]

colors = ['#E74C3C', '#2ECC71'] 

fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()

for i, feature in enumerate(features):
    counts = df[feature].value_counts().sort_index()  # ensure 0 then 1
    
    axes[i].pie(
        counts,
        labels=['No', 'Yes'],
        colors=colors,
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops={'edgecolor': 'white'}
    )
    axes[i].set_title(feature, fontsize=10)

plt.tight_layout()
plt.show()
```
